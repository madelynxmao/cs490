# Data Collection

```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import requests
from pprint import pprint
import finvizfinance 
import finvizfinance.screener 
from finvizfinance.quote import finvizfinance

%matplotlib inline
```

```python
# Input ticker you want to obtain news for
ticker = 'TSLA'

# Obtain news per ticker
stock = finvizfinance(ticker)
news = stock.ticker_news()

news['Ticker'] = ticker
news.head()
news['Date']= pd.to_datetime(news['Date'])
news.to_csv('../stock_news')
news.columns
```

```python
import yfinance as yf
import pandas as pd

# Define the stock ticker symbol and the date range

news['Date'] = pd.to_datetime(news['Date'])

# Get the earliest date in the DataFrame
start_date = news['Date'].min()
end_date = news['Date'].max()

# Fetch the stock data
stock_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
stock_data = stock_data.reset_index()
stock_data['Date'] = pd.to_datetime(stock_data['Date'])

# Ensure 'news['Date']' is in datetime format
news['Date'] = pd.to_datetime(news['Date']).dt.floor('d')  # Floor to the nearest day

# Merge the DataFrames on the 'Date' column
merged_data = pd.merge(news, stock_data, on='Date', how='inner')
```
